````python
def summarize_html_with_langchain_stuff(file_path, llm_model=LLM):
    print(f"\n--- Summarizing with 'stuff' method: {file_path} ---")
    loader = BSHTMLLoader(file_path) # Use BSHTMLLoader
    try:
        documents = loader.load() # BSHTMLLoader extracts text content
    except Exception as e:
        return f"Error loading HTML document: {e}"

    if not documents or not documents[0].page_content.strip():
        return "Error: No content loaded from the HTML document."

    print(f"   Loaded content length (chars): {len(documents[0].page_content)}")
    # Add token check if needed here

    chain = load_summarize_chain(llm_model, chain_type="stuff", verbose=False)
    try:
        summary_output = chain.invoke(documents)
        return summary_output['output_text']
    except Exception as e:
        return f"Error during LangChain 'stuff' summarization: {e}"
````

````python
def summarize_html_with_langchain_map_reduce(file_path, llm_model=LLM):
    print(f"\n--- Summarizing with 'map_reduce' method: {file_path} ---")
    loader = BSHTMLLoader(file_path) # Use BSHTMLLoader
    try:
        documents = loader.load()
    except Exception as e:
        return f"Error loading HTML document: {e}"

    if not documents or not documents[0].page_content.strip():
        return "Error: No content loaded from the HTML document."

    print(f"   Loaded content length (chars): {len(documents[0].page_content)}")

    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=10000, # Adjust based on LLM and content. Gemini 1.5 can handle larger.
        chunk_overlap=500,
        separators=["\n\n", "\n", " ", ""] # Good for extracted text
    )
    # documents from BSHTMLLoader is a list (usually with one Document for the whole HTML's text)
    # We split this Document object.
    split_docs = text_splitter.split_documents(documents)
    print(f"   Document split into {len(split_docs)} chunks for map_reduce.")

    if not split_docs:
        return "Error: Failed to split document into manageable chunks."

    chain = load_summarize_chain(llm_model, chain_type="map_reduce", verbose=False)
    try:
        summary_output = chain.invoke(split_docs)
        return summary_output['output_text']
    except Exception as e:
        return f"Error during LangChain 'map_reduce' summarization: {e}"
````

````python
class QueryType(JsonOutputParser):
    def get_format_instructions(self) -> str:
        return 'Return a JSON object with a single key "query_type" and one of the following values: "content_summary", "content_qa", "structure_code", "other".'
````

````python
router_prompt_template = """Classify the user's question into one of the following categories:
- "content_summary": If the question asks for a summary, overview, or general understanding of the document's textual content.
- "content_qa": If the question asks a specific factual question that can likely be answered from the document's textual content.
- "structure_code": If the question is about the HTML structure, specific HTML tags, or asks to write Python code to parse or extract data from the HTML.
- "other": If the question does not fit well into the above categories or is ambiguous.

**All questions should be accompanied by code that creates a visualization supporting the answer to the user's question**
- This code will fall into the structure code category even if the query is content_summary, content_qa, or other.

User Question: {question}

Classification (JSON with "query_type" key):
"""
router_prompt = PromptTemplate.from_template(router_prompt_template)
router_chain = router_prompt | LLM | QueryType() # QueryType is a JsonOutputParser
````

````python
content_qa_prompt_template = """You are a helpful AI Sports Data assistant. Answer the user's question based *only* on the provided Document Text.
If the information is not in the text, say so. Your job has two parts:
1) Answer the user's question based *only* on the provided Document Text;
2) Generate a Python code snippet (using matplotlib) that visualizes relevant data from the text or from your answer.

**Coding instructions**
You are an expert Python programmer, HTML analyst, and Sports Analyst.
Based *only* on the provided Raw HTML Source Code, answer the user's question or perform the requested task.
If writing Python code (e.g., using BeautifulSoup), provide only the Python code block enclosed in ```python ... ```.

**IMPORTANT INSTRUCTIONS:**
-   **DO NOT include the provided 'Raw HTML Source Code' as a string literal within the generated Python script.**
-   Assume the HTML content will be loaded into a variable named `html_content` within the Python script. Your script should operate on this `html_content` variable.
-   Provide only the Python code block enclosed in ```python ... ```.
-   The script should be runnable if the `html_content` variable is pre-populated with the HTML.
-   The script should run and create the visualization without any additions such as calling a function.
-   Use HTML_FILE_PATH as an existing variable for the HTML file, assume this variable is already set and does not need to be defined.
-   Use with open(HTML_FILE_PATH, encoding="utf-8") as f: soup = BeautifulSoup(f, "lxml") method to open HTML files.

**IMPORTANT INFORMATION:**
-   Pandas version being used is 2.1.4.
-   matplotlib version being used is 3.5.1.
-   BeautifulSoup version being used is 4.11.1.

Raw HTML Source Code (FOR YOUR REFERENCE ONLY, DO NOT INCLUDE IN SCRIPT):
---
{raw_html_source}
---

Document Text:
---
{document_text}
---

User Question: {question}

Content Answer:

Python Script (assuming `HTML_FILE_PATH` variable exists):"""
content_qa_prompt = PromptTemplate.from_template(content_qa_prompt_template)
content_qa_chain = (
    {
        "document_text": RunnableLambda(lambda x: extracted_text_content), # Always provide full extracted text
        "raw_html_source": RunnableLambda(lambda x: raw_html_content),
        "question": RunnablePassthrough() # Pass the original question
    }
    | content_qa_prompt
    | LLM
    | StrOutputParser()
)
````

````python
structure_code_prompt_template = """You are an expert Python programmer, HTML analyst, and Sports Analyst.
Based *only* on the provided Raw HTML Source Code, answer the user's question or perform the requested task.
If writing Python code (e.g., using BeautifulSoup), provide only the Python code block enclosed in ```python ... ```.

**IMPORTANT INSTRUCTIONS:**
-   **DO NOT include the provided 'Raw HTML Source Code' as a string literal within the generated Python script.**
-   Assume the HTML content will be loaded into a variable named `html_content` within the Python script. Your script should operate on this `html_content` variable.
-   Provide only the Python code block enclosed in ```python ... ```.
-   The script should be runnable if the `html_content` variable is pre-populated with the HTML.
-   The script should run and create the visualization without any additions such as calling a function.
-   Use HTML_FILE_PATH as an existing variable for the HTML file, assume this variable is already set and does not need to be defined.
-   Use with open(HTML_FILE_PATH, encoding="utf-8") as f: soup = BeautifulSoup(f, "lxml") method to open HTML files.

**IMPORTANT INFORMATION:**
-   Pandas version being used is 2.1.4.
-   matplotlib version being used is 3.5.1.
-   BeautifulSoup version being used is 4.11.1.

Raw HTML Source Code (FOR YOUR REFERENCE ONLY, DO NOT INCLUDE IN SCRIPT):
---
{raw_html_source}
---

User Question/Task: {question}

Python Script (assuming `HTML_FILE_PATH` variable exists):"""
structure_code_prompt = PromptTemplate.from_template(structure_code_prompt_template)
structure_code_chain = (
    {
        "raw_html_source": RunnableLambda(lambda x: raw_html_content), # Always provide full raw HTML
        "question": RunnablePassthrough() # Pass the original question
    }
    | structure_code_prompt
    | LLM
    | StrOutputParser()
)
````

````python
def ask_intelligent_assistant(user_question):
    print(f"\n❓ User Question: '{user_question}'")

    # Step 1: Route the question
    print("   Routing question...")
    try:
        route_result = router_chain.invoke({"question": user_question})
        query_type = route_result.get("query_type")
        print(f"   Determined query type: {query_type}")
    except Exception as e:
        print(f"   🔴 Error in routing: {e}. Defaulting to general approach.")
        query_type = "other" # Fallback

    # Step 2: Execute specialized chain
    response = ""
    if query_type in ["content_summary", "content_qa"]:
        print("   Processing as content question...")
        # Check if extracted_text_content + question + prompt is too long for context
        # (Add context window check here if necessary for this specific chain's input)
        response = content_qa_chain.invoke(user_question)
    elif query_type == "structure_code":
        print("   Processing as structure/code question...")
        # Check if raw_html_content + question + prompt is too long for context
        # (Add context window check here; raw HTML can be very large)
        response = structure_code_chain.invoke(user_question)
    else: # "other" or fallback
        print("   Processing as 'other' or fallback...")
        # For "other", you might have a more general prompt or ask for clarification
        # For simplicity, let's try a general prompt using extracted text
        general_prompt_text = f"Please try to answer the following question based on the document's content: {user_question}\n\nDocument Text:\n{extracted_text_content}"
        try:
            response = llm.invoke(general_prompt_text).content
        except Exception as e:
            response = f"Sorry, I encountered an error trying to process that: {e}"

    #print("\n💡 Assistant's Response:")
    #print(response)
    return response
````

````python
import re
def extract_code_blocks(markdown_text):
    code_blocks = []
    pattern = r"```(?P<language>\w*)\n(?P<code>[\s\S]*?)```"
    matches = re.finditer(pattern, markdown_text)
    for match in matches:
        language = match.group("language")
        code = match.group("code").strip()
        code_blocks.append({"language": language, "code": code})
    return code_blocks
````

````python
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

# Assume html_content is pre-populated with the HTML content
# Assume HTML_FILE_PATH is an existing variable for the HTML file

# Load HTML content from the file
with open(HTML_FILE_PATH, encoding="utf-8") as f:
    soup = BeautifulSoup(f, "lxml")

# Find the play-by-play table
pbp_table = soup.find("table", id="pbp")

if pbp_table:
    nyk_scores = []
    ind_scores = []
    play_indices = []
    current_nyk_score = -1 # Initialize with values that won't match first score
    current_ind_score = -1
    play_index_counter = 0 # This will be the x-axis for the plot

    # Iterate through table rows
    for row in pbp_table.find_all("tr"):
        # Skip header rows (including quarter headers and column headers)
        if "thead" in row.get("class", []):
            continue
        if row.find("th", colspan="6"): # Skip quarter headers like "1st Q"
            continue
        if row.find("th", text="Time"): # Skip the main column header row
            continue

        cells = row.find_all("td")
        if len(cells) >= 4: # Ensure there are enough columns for score
            score_cell = cells[3] # The score is in the 4th column (index 3)
            score_text = score_cell.get_text(strip=True)

            if '-' in score_text:
                try:
                    nyk_score_str, ind_score_str = score_text.split('-')
                    nyk_score = int(nyk_score_str)
                    ind_score = int(ind_score_str)

                    # Only record if the score actually changed or if it's the very first valid score
                    if nyk_score != current_nyk_score or ind_score != current_ind_score:
                        nyk_scores.append(nyk_score)
                        ind_scores.append(ind_score)
                        play_indices.append(play_index_counter)
                        current_nyk_score = nyk_score
                        current_ind_score = ind_score
                        play_index_counter += 1 # Increment only when a new score point is added
                except ValueError:
                    # This row's score cell doesn't contain a valid score format (e.g., "Jump ball:")
                    continue

    # Create the plot
    plt.figure(figsize=(12, 6))
    plt.plot(play_indices, nyk_scores, label='New York Knicks', color='blue', marker='o', markersize=4, linestyle='-')
    plt.plot(play_indices, ind_scores, label='Indiana Pacers', color='orange', marker='x', markersize=4, linestyle='--')

    plt.title('Score Progression: Knicks vs Pacers (May 27, 2025)')
    plt.xlabel('Event Number (Approximate)')
    plt.ylabel('Score')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
````

````python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
import re

# Assuming html_content is already loaded with the HTML source

#soup = BeautifulSoup(HTML_FILE_PATH, 'html.parser')
# Load the HTML file
with open(HTML_FILE_PATH, encoding="utf-8") as f:
    soup = BeautifulSoup(f, "lxml")
#print(soup)
# Find the play-by-play table
pbp_table = soup.find('table', {'id': 'pbp'})
#print(pbp_table)
if pbp_table:
    # Extract data from the table
    data = []
    rows = pbp_table.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        if len(cols) == 6:
            data.append(cols)

    # Create a Pandas DataFrame
    df = pd.DataFrame(data, columns=['Time', 'New York Play', 'NY Score Change', 'Score', 'IND Score Change', 'Indiana Play'])

    # Clean the score column and create NY and IND score columns
    def extract_scores(score_str):
        if score_str and re.match(r'\d+-\d+', score_str):
            ny_score, ind_score = map(int, score_str.split('-'))
            return ny_score, ind_score
        else:
            return None, None

    df[['New York Score', 'Indiana Score']] = df['Score'].apply(lambda x: pd.Series(extract_scores(x)))

    df = df.dropna(subset=['New York Score', 'Indiana Score'], how='any')
    df['New York Score'] = df['New York Score'].astype(int)
    df['Indiana Score'] = df['Indiana Score'].astype(int)

    # Create the score difference column
    df['Score Difference'] = df['Indiana Score'] - df['New York Score']

    # Plotting the game flow
    plt.figure(figsize=(12, 6))
    plt.plot(df.index.values, df['Score Difference'].values, label='Score Difference (IND - NYK)')
    plt.xlabel('Play Index')
    plt.ylabel('Score Difference')
    plt.title('Game Flow: Knicks vs Pacers')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Save the plot to a file (optional)
    plt.savefig('game_flow.png')

    # Display the plot (optional - if running in an environment that supports it)
    plt.show()

else:
    print("Play-by-play table not found.")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

# Load the HTML content into BeautifulSoup
with open(HTML_FILE_PATH, encoding="utf-8") as f:
    soup = BeautifulSoup(f, "lxml")

# Find the play-by-play table
table = soup.find('table', {'id': 'pbp'})

# Extract the data from the table
data = []
rows = table.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Clean the DataFrame
df = df.dropna(how='all')
df = df[df[0] != 'Time']
df = df[df[0] != 'End of 1st quarter']
df = df[df[0] != 'End of 2nd quarter']
df = df[df[0] != 'End of 3rd quarter']
df = df[df[0] != 'End of 4th quarter']

# Rename the columns
df.columns = ['Time', 'New York Play', 'New York Score Change', 'Score', 'Indiana Score Change', 'Indiana Play']

# Extract the scores for each team
scores = df['Score'].str.split('-', expand=True)
scores.columns = ['New York Score', 'Indiana Score']

# Concatenate the scores with the original DataFrame
df = pd.concat([df, scores], axis=1)

# Convert the scores to numeric values
df['New York Score'] = pd.to_numeric(df['New York Score'], errors='coerce')
df['Indiana Score'] = pd.to_numeric(df['Indiana Score'], errors='coerce')

# Forward fill missing values to get cumulative scores
df['New York Score'] = df['New York Score'].ffill()
df['Indiana Score'] = df['Indiana Score'].ffill()

# Convert time to minutes remaining
def time_to_minutes_remaining(time_str, quarter):
    minutes, seconds = map(int, time_str.split(':'))
    total_seconds = minutes * 60 + seconds
    minutes_remaining = 12 * (4 - quarter) + total_seconds / 60
    return minutes_remaining

# Create a list of minutes remaining for each play
minutes_remaining = []
quarter = 1
for i, row in df.iterrows():
    time_str = row['Time']
    if '2nd Q' in row['Time']:
        quarter = 2
        continue
    elif '3rd Q' in row['Time']:
        quarter = 3
        continue
    elif '4th Q' in row['Time']:
        quarter = 4
        continue
    minutes_remaining.append(time_to_minutes_remaining(time_str, quarter))

# Assign the minutes remaining to the DataFrame
df['Minutes Remaining'] = minutes_remaining

# Create the gameflow visualization
plt.figure(figsize=(12, 6))
plt.plot(df['Minutes Remaining'], df['New York Score'], label='New York Knicks', color='blue')
plt.plot(df['Minutes Remaining'], df['Indiana Score'], label='Indiana Pacers', color='gold')

# Add labels and title
plt.xlabel('Minutes Remaining')
plt.ylabel('Score')
plt.title('Gameflow: New York Knicks vs Indiana Pacers')
plt.xlim(max(df['Minutes Remaining']), 0)  # Invert x-axis
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
````

````python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

# Load the HTML file
with open(HTML_FILE_PATH, encoding="utf-8") as f:
    soup = BeautifulSoup(f, "lxml")

# Locate the play-by-play table
pbp_table = soup.find("table", id="pbp")

# Initialize data structures
times = []
nyk_scores = []
ind_scores = []

# Starting scores
nyk_score = 0
ind_score = 0

# Loop through play-by-play rows
for row in pbp_table.find_all("tr"):
    cells = row.find_all("td")
    if len(cells) == 6:
        time_cell = cells[0].text.strip()
        score_cell = cells[3].text.strip()

        # If score updated, extract numeric values
        if score_cell and '-' in score_cell:
            nyk_score, ind_score = map(int, score_cell.split('-'))

            # Append time (convert to game clock seconds) and current scores
            # Example: "11:36.0" → 12*60 - 11*60 - 36 = 24 seconds elapsed
            min_sec = time_cell.split(":")
            if len(min_sec) == 2:
                minutes, seconds = map(float, min_sec)
                game_clock = (12 * 60) - (minutes * 60 + seconds)
            else:
                game_clock = None  # For jump ball or non-time rows

            if game_clock is not None:
                times.append(game_clock)
                nyk_scores.append(nyk_score)
                ind_scores.append(ind_score)

# Reverse time axis so it runs from 0 to end of game
times = [12*60 - t for t in times]

# Plot game flow
plt.figure(figsize=(12, 6))
plt.plot(times, nyk_scores, label="New York Knicks", color="orange", linewidth=2)
plt.plot(times, ind_scores, label="Indiana Pacers", color="royalblue", linewidth=2)

plt.xlabel("Game Time (seconds elapsed)")
plt.ylabel("Score")
plt.title("Game Flow: Knicks vs Pacers, May 27, 2025")
plt.legend()
plt.grid(True)
plt.gca().invert_xaxis()  # Show time counting down
plt.tight_layout()
plt.show()
````
